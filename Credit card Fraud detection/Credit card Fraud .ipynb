```python
#Transaction count for two day in hour basis
plt.figure(figsize = (10,4))
sns.countplot(x = 'Hour',data = Data,hue = 'Hour', palette = 'viridis',legend=False)
plt.title('Transaction count by Hour of the day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

```python
Data['Hourofday'] = (Data['Time']//3600%24).astype(int)
```

```python
#Transactions of one day
plt.figure(figsize=(10,4))
sns.countplot(x = 'Hourofday',data = Data,hue='Hourofday', palette = 'viridis',legend=False)
plt.title('Transaction count by hour of day')
plt.xlabel('Hourofday')
plt.ylabel('Count')
plt.show()
```

```python
#Transaction count for two days

Data['Day']= (Data['Time']//86400).astype(int)
Data['Hour']= (Data['Time']//3600%24).astype(int)

plt.figure(figsize=(10,5))
sns.countplot(x = 'Hour',data= Data,hue='Day',palette='viridis',legend=False)
plt.title('Transaction for two days')
plt.xlabel('Hourofday')
plt.ylabel('Count')
plt.show()
```

```python
#Transactions line chart

Hourly = Data.groupby(['Day','Hour']).size().reset_index(name='Count')

plt.figure(figsize=(10,5))
sns.lineplot(x = 'Hour', y = 'Count',hue = 'Day', data= Hourly, marker = 'o')
plt.title('Transactions trend')
plt.xlabel('Hourofday')
plt.ylabel('Count')
plt.show()
```

```python
#Comparison between Normal vs Fraud transactions

#Lets see how inmalanced the data is

sns.countplot(x = 'Class', data= Data)
plt.title('Normal vs Fraud')
plt.xlabel('Transaction type (0 = Normal,1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
#lets compare when at what time frauds occur frequently

plt.figure(figsize=(10,5))
sns.histplot(x = 'Hour',data=Data, hue = 'Class', multiple = 'stack', bins = 24)
plt.title('Transactions(Normal vs Fraud)')
plt.xlabel('Hour of day')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x= 'Hour', data = Data, hue = 'Class', stat='percent',multiple = 'fill', bins = 24)
plt.title('percente of Fraud vs Normal transactions per hour')
plt.xlabel('Hour')
plt.ylabel('Percent')
plt.show()

#Orage bar shows percent of fraud
```

```python
#Now we will highlight only fraud transactions 

Fraud = (Data[Data['Class']==1]).copy()
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data = Fraud, color = 'red')
plt.title('Fraud Transactions by Hour')
plt.xlabel('Hour')
plt.ylabel('Fraud count')
plt.show()
```

```python
#Fraud for two days
Fraud['Hour'] = Fraud['Time']//3600
Fraud['Day']= (Fraud['Hour']//24)+1
Fraud['Hourofday'] =Fraud ['Hour']%24

plt.figure(figsize=(10,5))
sns.histplot(x = 'Hourofday', data=Fraud , hue='Day', bins = 24, multiple = 'dodge', palette = 'viridis')
plt.title('Fraud transactions by hour (Two days)')
plt.xlabel('Hour of day')
plt.ylabel('Fraud count')
plt.show()
```
